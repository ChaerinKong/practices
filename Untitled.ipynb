```python
import pandas as pd
kospi = pd.read_csv('kospi200.csv')
kospi.head()
```

```python
kospi.dtypes
```

```python
kospi['Date'] = pd.to_datetime(kospi['Date'])
```

```python
kospi['Date'].head()
```

```python
kospi['KOSPI'].head()
```

```python
kospi['KOSPI'].tail()
```

```python
kospi = kospi.sort_values(by='Date', ascending=False)
```

```python
kospi.head()
```

```python
kospi.head()
```

```python
kospi.tail()
```

```python
kospi['by_month'] = kospi['Date'].dt.to_period('m')
```

```python
kospi.head()
```

```python
kospi.tail()
```

```python
kospi_by_month = kospi.groupby('by_month').head(1)
```

```python
kospi_by_month
```

```python
kospi_by_month
```

```python
kospi_by_month = kospi_by_month.drop(columns='by_month')
```

```python
kospi_by_month.to_csv('KOSPI_data.csv', index=False)
```
